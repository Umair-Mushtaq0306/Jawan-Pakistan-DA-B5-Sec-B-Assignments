Q10. Set Comparison
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7}

Find common elements.
Find elements only in Set A.
Find elements only in Set B.

```python
A = {1, 2, 3, 4, 5} 
B = {4, 5, 6, 7}
```

```python
A.intersection(B)
```

```python
A.difference(B)
```

```python
B.difference(A)
```

```python
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7}

# Find common elements (intersection)
common = A.intersection(B)
print("Common elements:", common)

# Find elements only in Set A (difference)
only_in_A = A.difference(B)
print("Elements only in Set A:", only_in_A)

# Find elements only in Set B (difference)
only_in_B = B.difference(A)
print("Elements only in Set B:", only_in_B)
```
